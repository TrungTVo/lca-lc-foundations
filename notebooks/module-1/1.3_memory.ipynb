```python
from dotenv import load_dotenv
from pprint import pprint
from langchain.agents import create_agent
from langchain.messages import HumanMessage

load_dotenv()
```

## No memory

```python
agent = create_agent(
    "gpt-5-nano"
)
```

```python
question = HumanMessage(content="Hello my name is Seán and my favourite colour is green")

response = agent.invoke(
    input={"messages": [question]}
)
```

```python
pprint(response)
```

```python
question = HumanMessage(content="What's my favourite colour?")

response = agent.invoke(
    {"messages": [question]} 
)

pprint(response)
```

## Config with In-Memory Saver

```python
from langgraph.checkpoint.memory import InMemorySaver

agent = create_agent(
    model="gpt-5-nano",
    checkpointer=InMemorySaver()
)
```

### Conversation for User 1

```python
question = HumanMessage(content="Hello my name is Seán and my favourite colour is green")
config = {
    "configurable": {"thread_id": "session-1"}
}

response = agent.invoke(
    input={"messages": [question]},
    config=config
)
```

```python
pprint(response)
```

```python
question = HumanMessage(content="What's my favourite colour?")

response = agent.invoke(
    input={"messages": [question]},
    config=config
)

pprint(response)
```

### Conversation for User 2

```python
config2 = {
    "configurable": {"thread_id": "session-2"}
}

response = agent.invoke(
    input={"messages": [
        HumanMessage(content="Hello my name is Trung and my favourite colour is red")
    ]},
    config=config2
)
```

```python
response = agent.invoke(
    input={"messages": [
        HumanMessage(content="What's my favourite colour?")
    ]},
    config=config2
)
```

```python
pprint(response)
```

## Get the full saved state for a thread

### Get full saved state for thread `"session-1"` (user 1)

```python
state = agent.get_state(config=config)
pprint(state.values["messages"])  # full message history
```

### Get full saved state for thread `"session-2"` (user 2)

```python
state = agent.get_state(config=config2)
pprint(state.values["messages"])  # full message history
```
